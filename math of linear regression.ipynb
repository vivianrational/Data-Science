## Simple linear regression
The formula of linear regression is$\hat{y}=E[Y|X=x]=\alpha+\beta x$ 
<br>in a common case, we usually see $Y=\alpha +\beta x+\epsilon$ 
We have many methods to find $\beta$,such as Gradient Descent and Least Squares.
<br> Normally, we choose to use Least Squares method 
$\beta = (X^TX)^{-1}X^Ty$

### The Least Squares Assumptions
Personally, I think these assumptions are very important.
- 1) The regression model is linear in the coefficients and the error term
- 2) The conditional mean should be zero. 
<br>E[$\epsilon$|X]=0 implies that X and $\epsilon$ are uncorrelated.
- 3)  $(X_i,Y_i)$ are independently and identically distributed
- 4) Large outliers are unlikely.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 20)
y=4*x
y1=3*x
p1=np.poly1d(np.polyfit(x,y1,1))
p=np.poly1d(np.polyfit(x,y,1))
t= np.linspace(0, 1, 20)
plt.plot(0.8,3.2,"o",0.85,0.85*3,"x",t,p(t),t,p1(t),'-')

plt.legend(["Data plane","True plane"])
```

In the case of least squares approximation, we are minimizing the Euclidean distance.
You could imagine the blue dot is our observations, and then we need to project these data into the red line plane.
